```python
import pandas as pd
import os
import glob
from datetime import datetime
from sqlalchemy import create_engine
#from geopy.distance import distance
```

```python
# get the old format csv files 201901 - 202101
old_path = 'Resources/oldFormat'
old_bike_history = pd.DataFrame()
old_csv_files = glob.glob(os.path.join(old_path, "*.csv"))
```

```python

# Loop through files to store all csvs in a df
for f in old_csv_files:
    # read the file into dataframe
    file_load = pd.read_csv(f)
    
    # add it to the main dataframe 
    old_bike_history = old_bike_history.append(file_load, ignore_index=True)
        
    # print success
    # print(f"loaded {f}")
```

```python
old_bike_history.head()
```

```python
old_bike_history_filtered =  old_bike_history[  ["starttime", "stoptime", "start station id", "start station name", 
                                                 "start station latitude", "start station longitude", "end station id",
                                                 "end station name", "end station latitude", "end station longitude", "usertype"
                                                ]   ]
```

```python
old_bike_history_filtered.head()
```

```python
old_bike_history_filt_rename = old_bike_history_filtered.rename(columns={
    "starttime": "started_at",
    "stoptime": "ended_at",
    "start station id": "start_station_id",
    "start station name": "start_station_name",
    "start station latitude": "start_lat",
    "start station longitude": "start_lng",
    "end station id": "end_station_id",
    "end station name": "end_station_name",
    "end station latitude": "end_lat",
    "end station longitude": "end_lng"

})

```

```python
old_bike_history_filt_rename.head()
```

```python
# get the new format csv files 202101 - 202112
new_path = 'Resources/newFormat'
new_bike_history = pd.DataFrame()
new_csv_files = glob.glob(os.path.join(new_path, "*.csv"))
```

```python
# Loop through files to store all csvs in a df
for f in new_csv_files:
    # read the file into dataframe
    file_load = pd.read_csv(f)
    
    # add it to the main dataframe 
    new_bike_history = new_bike_history.append(file_load, ignore_index=True)
        
    # print success
    # print(f"loaded {f}")
```

```python
new_bike_history.head()
```

```python
member_casual = new_bike_history["member_casual"]
```

```python
# member_casual
```

```python
#type(member_casual)
```

```python
usertype_new = []
```

```python
for i in member_casual:
    if i == "member":
        usertype_new.append("Subscriber")
    elif i == "casual":
        usertype_new.append("Customer")
    else:
        usertype_new.append("unknown")
        
```

```python
# usertype_new
```

```python
new_bike_history["usertype"] = usertype_new
```

```python
new_bike_history.head()
```

```python
new_bike_history_filtered =  new_bike_history[  ["started_at", "ended_at", "start_station_id", "start_station_name", 
                                                 "start_lat", "start_lng", "end_station_id",
                                                 "end_station_name", "end_lat", "end_lng", "usertype"
                                                ]   ]
```

```python
new_bike_history_filtered.head()
```

```python
#Do a Union between the two daataframes
bike_history_combined = pd.concat([old_bike_history_filt_rename, new_bike_history_filtered], ignore_index=True)
bike_history_combined
```

```python
bike_history_combined.dtypes
```

```python
bike_history_combined["started_at"] = pd.to_datetime(bike_history_combined["started_at"])
```

```python
bike_history_combined["ended_at"] = pd.to_datetime(bike_history_combined["ended_at"])
```

```python
bike_history_combined.dtypes
```

```python
bike_history_combined.shape
```

```python
bike_history_combined = bike_history_combined.dropna()
bike_history_combined.shape
```

```python
# Calculate the duration for the arrival after the incident was reported

bike_history_combined["trip_duration"] = (bike_history_combined["ended_at"] - bike_history_combined["started_at"]).astype('timedelta64[m]')
```

```python
bike_history_combined.head()
```

```python
bike_history_combined.shape
```

```python
only_wythe = bike_history_combined.loc[  (bike_history_combined["start_station_name"] == "Wythe Ave & Metropolitan Ave") |  (bike_history_combined["end_station_name"] == "Wythe Ave & Metropolitan Ave"), :]
```

```python
only_wythe
```

```python
# Drop the rows where the arrival duration column is less than 0
bike_history_combined = bike_history_combined.drop(bike_history_combined[bike_history_combined["trip_duration"] < 0 ].index)
bike_history_combined.shape
```

```python
# drop dupliacte rows 
bike_history_combined = bike_history_combined.drop_duplicates()
bike_history_combined.shape
```

```python
# Calculate the distance in miles for each trip

# bike_history_combined["trip_miles"] = geopy.distance.vincenty((start_lat,start_lng), (end_lat,end_lng)).miles
```

```python
# Add a month variable in the form of MM/1/YYYY
bike_history_combined["start_month"] = pd.to_datetime(bike_history_combined["started_at"]).astype('datetime64[M]')
```
